```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import (StaleElementReferenceException,NoSuchElementException)

driverLoc = 'D:\\GitHub\\chromedriver\\chromedriver.exe'
options = webdriver.ChromeOptions()

options.add_argument("--headless")
options.add_argument("--disable-gpu")

recipe_info = []
```

```python
import time
driver = webdriver.Chrome(driverLoc, chrome_options=options)
#url = 'https://www.foodsafetykorea.go.kr/portal/cookrcp/cookRcpInfoList.do?rcp_cat=RC_COOK02&menu_no=3608&menu_grp=MENU_NEW03' #저당레시피
url = 'https://www.foodsafetykorea.go.kr/portal/cookrcp/cookRcpInfoList.do?rcp_cat=RC_COOK03&menu_no=3608&menu_grp=MENU_NEW03' #삼삼한밥상
driver.get(url)
driver.implicitly_wait(2)
wait = WebDriverWait(driver, 10)

driver.execute_script("arguments[0].click();", driver.find_element(By.XPATH, '//*[@id="rcp_pat"]/option[1]'))
driver.execute_script("arguments[0].click();", driver.find_element(By.XPATH, '//*[@id="btnSearch"]'))
driver.implicitly_wait(1)
wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="searchRltFrame"]/div[1]/div/ul[1]/li[2]/a'))).click()
driver.find_element(by=By.XPATH, value='//*[@id="boardFooter"]/a').click()

for _ in range(38):
    driver.execute_script("arguments[0].click();", driver.find_element(By.XPATH, '//*[@id="boardFooter"]/div/ul/li[7]/a'))
    time.sleep(1)

l=39
while l <= 59: #59 8
    for i in range(1, 5):
        for r in range(1, 5):
            print(l,i,r)
            try:
                driver.implicitly_wait(3)
                element_xpath = f'//*[@id="searchRltFrame"]/div[{i}]/div/ul[{r}]/li[2]/a'
                driver.execute_script("arguments[0].click();", driver.find_element(By.XPATH, element_xpath))
                wait.until(EC.presence_of_element_located((By.CLASS_NAME, "section")))

                page_html = BeautifulSoup(driver.page_source, "html.parser")
                body = page_html.find("div", class_="section")

                title = page_html.find("div", class_="recipe-head").text.strip()
                ingredients = body.find("div", class_='sr material_tit').text.strip().replace("\n", '').replace("\t", '').replace('재료', '')
                methods = body.find("div", class_='col-3').text.strip().split('\n\n\n\n\n\n')
                nutrients = body.find("div", class_='sl ingredient').text.strip().replace(
                    "영양성분\n\n영양성분 정보(열량, 탄수화물, 단백질, 지방, 나트륨)를 제공하는 표 입니다.", '').replace(
                    '\t', '').replace('\n', '').replace(
                    "열량탄수화물단백질지방나트륨",'') #열량탄수화물단백질지방나트륨 열량탄수화물단백질지방당

                recipe_info.append({'Title': title, 'Ingredients': ingredients, 'Methods': methods, 'Nutrients': nutrients})
                driver.execute_script("arguments[0].click();", driver.find_element(By.XPATH, '//*[@id="boardFooter"]/a'))
                if i == 4 and r == 4:
                    driver.execute_script("arguments[0].click();", driver.find_element(By.XPATH, '//*[@id="boardFooter"]/div/ul/li[7]/a'))
                    l += 1
            except (StaleElementReferenceException, NoSuchElementException) as e:
                print(f"Exception occurred: {e}")
                i=1
                break
driver.quit()
```

```python
dang_df = pd.DataFrame(recipe_info)
dang_df.columns= ['레시피명','재료','방법','영양성분']
dang_df = dang_df.drop_duplicates(subset=['레시피명'])
dang_df['열량'] = dang_df['영양성분'].str.extract(r'열량(.*?)kcal탄수화물').astype(float)
dang_df['탄수화물'] = dang_df['영양성분'].str.extract(r'탄수화물(.*?)g단백질').astype(float)
dang_df['단백질'] = dang_df['영양성분'].str.extract(r'단백질(.*?)g지방').astype(float)
#dang_df['지방'] = dang_df['영양성분'].str.extract(r'지방(.*?)g당').astype(float)
#dang_df['당'] = dang_df['영양성분'].str.extract(r'당(.*?)g').astype(float)
dang_df['지방'] = dang_df['영양성분'].str.extract(r'지방(.*?)g나트륨').astype(float)
dang_df['나트륨'] = dang_df['영양성분'].str.extract(r'나트륨(.*?)mg')
dang_df = dang_df.drop('영양성분', axis=1)
#dang_df = dang_df.drop_duplicates()
#dang_df.to_csv('저당레시피.csv',encoding='ANSI')
dang_df.to_csv('삼삼한밥상3.csv')
```
